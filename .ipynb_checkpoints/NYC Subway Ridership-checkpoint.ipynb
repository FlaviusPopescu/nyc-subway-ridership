We could also try to come up with a prediction model that, given a time of day and weather data, it can output the number of expected subway riders.

This dataset includes 207 stations (almost half of the total across all of New York City's boroughs) and 240 turnstile units and was collected during the month of May, 2011. There are more units than stations since some stations are larger (like connection hubs) and contain more units which have different IDs. That's why in the listing above some stations appear many times.

Now let's move on to the interesting part: subway ridership! Let's take a look at the total number of entries, across all stations/units, aggregated at each unique timestamp in our datasets. The improved dataset used in this analysis already has accummulated rows with different timestamps, such that most days will have 6 readings, every 4 hours starting with midnight.

To perform the aggregation, we can use *pandasql* and write a SQL query:

```python
query = """select 
             datetime as 'datetime',
             sum(ENTRIESn_hourly) as 'entries'
           from df
           group by datetime"""

from pandasql import *
import datetime
from ggplot import *
%matplotlib inline

# fp: make sure that when converting string dates to date objects
# we don't offset them to a local timezone and display them as they are
os.environ['TZ'] = "UTC"

# fp: convert the datetime column to a date object, since it will be used in ggplot
# as a date axis
total_entries = sqldf(query, locals())
total_entries['datetime'] = total_entries['datetime'].astype('datetime64[s]')
```

What does this dataframe look like? Let's use *ggplot* for that. Let's display the days of the week on the x-axis which we can since we have a datetime column type:

```python
ggplot(total_entries, aes('datetime', 'entries')) + geom_line(colour="steelblue") \
        + scale_x_date(breaks=date_breaks('2 days'), labels='%a') \
        + scale_y_continuous(labels='comma') \
        + ggtitle("Total entries to all stations during May 2011") \
        + xlab("Day of Week") + ylab("Total Entries")
```

Let's take a closer look at how a week looks and also plot the total exits.

```python
query = """select 
             datetime as 'datetime',
             sum(ENTRIESn_hourly) as 'entries'
           from df
           where datetime <= "2011-05-09 00:00:00"
           group by datetime"""

# entries in first week of May only
total_entries = sqldf(query, locals())
total_entries['datetime'] = total_entries['datetime'].astype('datetime64[s]')


query = """select 
             datetime as 'datetime',
             sum(EXITSn_hourly) as 'exits'
           from df
           where datetime <= "2011-05-09 00:00:00"
           group by datetime"""
# exits in first week of May only
total_exits = sqldf(query, locals())
total_exits['datetime'] = total_exits['datetime'].astype('datetime64[s]')

ggplot(total_entries, aes('datetime', 'entries')) + geom_line(colour="steelblue") \
        + scale_x_date(breaks=date_breaks('1 day'), labels='%a') \
        + scale_y_continuous(labels='comma') \
        + ggtitle("Total entries to all stations during first week of May 2011") \
        + xlab("Day of Week") + ylab("Total Entries") \
        + geom_line(aes('datetime', 'exits'), data=total_exits, colour="darkgreen")
        
```

Let's take a week day:

```python
query = """select 
             datetime as 'datetime',
             sum(ENTRIESn_hourly) as 'entries'
           from df
           where datetime between "2011-05-05 00:00:00" and "2011-05-06 00:00:00"
           group by datetime"""

# entries in first week of May only
total_entries = sqldf(query, locals())
total_entries['datetime'] = total_entries['datetime'].astype('datetime64[s]')

query = """select 
             datetime as 'datetime',
             sum(EXITSn_hourly) as 'exits'
           from df
           where datetime between "2011-05-05 00:00:00" and "2011-05-06 00:00:00"
           group by datetime"""
# exits in first week of May only
total_exits = sqldf(query, locals())
total_exits['datetime'] = total_exits['datetime'].astype('datetime64[s]')


ggplot(total_entries, aes('datetime', 'entries')) + geom_line(colour="steelblue") \
        + scale_x_date(breaks=date_breaks('2 hours'), labels='%H:%M') \
        + scale_y_continuous(labels='comma') \
        + ggtitle("Total entries to all stations during first week of May 2011") \
        + xlab("Day of Week") + ylab("Total Entries") \
        + geom_line(aes('datetime', 'exits'), data=total_exits, colour="darkgreen")
```

### Section 1 - Statistical Test

To analyze the subway data we can use the Mann-Whitney U-test. In Python, we can invoke it by usint the *scipy.stats* module:

    scipy.stats.mannwhitneyu()
    
and provide two data sets. So let's first split the original data into rainy/not rainy days and then pass these to the *mannwhitneyu* function:

```python
import numpy as np
import scipy

rainy_days = df[df.rain == 1]['ENTRIESn_hourly']
mean_entries_rainy = np.mean(rainy_days)

not_rainy_days = df[df.rain == 0]['ENTRIESn_hourly']
mean_entries_not_rainy = np.mean(not_rainy_days)

U, p = scipy.stats.mannwhitneyu(not_rainy_days, rainy_days)

print U, p
print mean_entries_rainy, mean_entries_not_rainy
```

The variable *p* now holds the one-sided p-value (as described in [the documentation](!http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html)).

The null hypothesis is that the two populations, rainy vs not-rain, come from the same distribution
